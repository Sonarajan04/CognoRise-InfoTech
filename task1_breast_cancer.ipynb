# balancing the dataset

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=1)
X_over,y_over=smote.fit_resample(X,y)
```

# scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X_over)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y_over,random_state=1,test_size=0.3)
```

# model creation

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
knn=KNeighborsClassifier()
rfc=RandomForestClassifier()
lr=LogisticRegression()
dt=DecisionTreeClassifier()
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
lst1=[knn,rfc,lr,dt]
for i in lst1:
  print('***********',i,'***********')
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(classification_report(y_test,y_pred))
```

### hyperparameter tunning on knn

```python
from sklearn.model_selection import GridSearchCV
param={'n_neighbors':[3,5,7,9],
       'weights':['uniform','distance'],
       'metric':['euclidian','manhattan']}
grid=GridSearchCV(knn,param,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
print(grid.best_params_)
```

```python
knn1=KNeighborsClassifier(n_neighbors=5,metric='manhattan',weights='uniform')
knn1.fit(X_train,y_train)
y_pred=knn1.predict(X_test)
print(classification_report(y_test,y_pred))
```
